# Iris dataset testing

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris

from dimensionality_reduction.alghoritms.reduction.pca import PCA
```

### Load iris

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

### PCA

```python
pca = PCA(n_components=2)
X_reduced_pca = pca.fit_transform(X)
print(f"Explained variance ratio: {dict(zip(iris.feature_names,
                                             pca.explained_variance_ratio_,
                                             strict=False))}")
print(f"Components shape: {pca.components_.shape}")
```

```python
df = pd.DataFrame(X_reduced_pca, columns=['PC1', 'PC2'])
df['species'] = [target_names[i] for i in y]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PC1', y='PC2', hue='species', palette='viridis', s=100)
plt.title('PCA ')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
